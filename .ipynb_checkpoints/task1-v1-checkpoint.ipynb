```python
from qiskit import *
from qiskit.quantum_info import random_statevector
from scipy.optimize import minimize
from scipy.optimize import Bounds
import numpy as np
import matplotlib.pyplot as plt
from qiskit.visualization import plot_histogram
import time
pi=np.pi
%matplotlib inline
backend = Aer.get_backend('statevector_simulator')
```

```python
def norm(vector):
    return sum(abs(vector)**2)
```

```python
def make_circuit(thetas,qr_length):
    # A set of theta parameters and number of qubits fully characterize the circuit.
    # Given compatible, the number of layers (L) can be obtained from the inputs.
    qr=QuantumRegister(qr_length,'q')
    U_total=QuantumCircuit(qr)
    L=len(thetas)/(2*qr_length)
    for cL in range(int(L)):
        # In each iteration, one new layer (U_block_cL) composed of its odd and even blocks is built 
        # and appended to the previous layers.
        theta_cL = thetas[int(cL*2*qr_length):int((cL+1)*2*qr_length)]
        theta_odd_cL  = theta_cL[0:int(qr_length)]
        theta_even_cL = theta_cL[int(qr_length):int(2*qr_length)]
        
        U_odd_cL=QuantumCircuit(qr)
        for jo in range(qr_length):
            U_odd_cL.rx(theta_odd_cL[jo],jo)
        
        U_even_cL=QuantumCircuit(qr)
        for je in range(qr_length):
            U_even_cL.rz(theta_even_cL[je],je)
        for je in range(qr_length-1):
            U_even_cL.cz(je,range(je+1,qr_length))
                
        U_block_cL=U_odd_cL+U_even_cL
        U_total=U_total+U_block_cL
        U_total.barrier()
        
    return U_total
```

```python
def _cost(thetas,qr_length,target_state):
    
    # The cost function, that is to be minimized in the theta space, takes as input the parameters to build the circuit
    # and a fixed target state.
    # The final state is the state obtained by evolving |00..0> according to the built circuit.
    
    U_total=make_circuit(thetas,qr_length)
    
    job=execute(U_total,backend)
    result = job.result()
    final_state=result.get_statevector()
    
    cost=norm(final_state-target_state.data)
    return cost
```

```python
# Definitions

qr_length=4 # Number of qubits for this task
dim=2**qr_length # dimension of state space
target_state=random_statevector(dim)
```

```python
# Main Code

maxL=10 # Maximum number of layers to seek for convergence
theta_set = np.ndarray(0)
cost=[]
for L in range(maxL):
    start=time.time()
    # In every iteration a new set of parameters are randomly generated 
    # and appended to the set of parameters that are optimized in the previous iteration
    new_set   = np.random.rand(2*qr_length)*2*pi 
    theta_set = np.append(theta_set,new_set)
    res=minimize(_cost,theta_set,args=(qr_length,target_state),bounds=Bounds(np.zeros(len(theta_set)),np.ones(len(theta_set))*2*pi),method='Powell')
    print(res.success) # prints "True" if minimization is successful
    theta_set=res.x # updating set of parameters
    cost.append(res.fun) # appends the last value of cost function to the list of costs to be plot later
    print('cost vector is \n',cost)
    end=time.time()
    print('time elapsed for L=',L+1,'is',end-start)
    if (L>3 and cost[L-1]<1e-6 and cost[L]<1e-6):   # check for two consecutive convergences to break iteration and keep number of layers as small as possible   
        maxL=L+1
        print('The required number of layers to converge is L = ',maxL)
        break
```

```python
# Plotting the cost function vs number of layers
plt.plot(range(1,maxL+1),cost)
plt.xlabel('number of layers')
plt.ylabel('cost')
plt.show()
```

```python
# Building the final circuit using the optimized parameters to collect measurement statistics
meas = QuantumCircuit(QuantumRegister(qr_length,'q'),ClassicalRegister(qr_length,'c0'))
meas.measure(range(qr_length), range(qr_length))
final_circ = make_circuit(theta_set,qr_length)+meas
# final_circ.draw('mpl')
```

```python
# comparing the operationally available probabilities from the simulated and target states
backend_sim = Aer.get_backend('qasm_simulator')
job_sim = execute(final_circ, backend_sim, shots=1024)
result_sim = job_sim.result()
counts = result_sim.get_counts()
plot_histogram([counts,target_state.sample_counts(1024)],figsize=(15,7),legend=['simulated','target'])
```
